#**Scraping Data:**
* The code uses BeautifulSoup and Selenium to scrape job information from the
Naukri.com website.
* It extracts job titles, company names, experience requirements, locations, and salary information.

```python
# Importing necessary libraries
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
import time
import pandas as pd
```

```python
# Sending a GET request to the Naukri website
res = requests.get('https://www.naukri.com/data-science-jobs?k=data%20science')
```

```python
# Opening a Chrome browser window using Selenium WebDriver
driver = webdriver.Chrome()
```

```python
# Accessing the specified URL using the WebDriver
# Pausing the execution for 3 seconds to allow the page to load
# Creating a BeautifulSoup object to parse the HTML content of the page
driver.get('https://www.naukri.com/data-science-jobs?k=data%20science&nignbevent_src=jobsearchDeskGNB')
time.sleep(3)
soup = BeautifulSoup(driver.page_source,'html.parser')
```

```python
soup
```

```python
# Extracting all anchor tags with the class 'title' from the BeautifulSoup object
soup1 = soup.find_all('a' ,class_ = 'title')
```

```python
# Extracting text content from each element in the list of anchor tags
list1 = [i.text for i in soup1]
```

```python
list1
```

```python
# Extracting all anchor tags with the class 'comp-name mw-25' from the BeautifulSoup object
soup2 = soup.find_all('a',class_ = 'comp-name mw-25')
```

```python
# Extracting text content from each element in the list of anchor tags
list2 = [i.text for i in soup2]
```

```python
list2
```

```python
# Extracting all span tags with the class 'expwdth' from the BeautifulSoup object
soup3 = soup.find_all('span', class_ = 'expwdth')
```

```python
# Extracting text content from each element in the list of span tags
list3 = [i.text for i in soup3]
```

```python
# Extracting text content from each element in the list of span tags with class 'locWdth'
soup4 = soup.find_all('span',class_ = 'locWdth')
```

```python
list4 = [i.text for i in soup4]
```

```python
# Extracting text content from each element in the list of span tags with class 'ni-job-tuple-icon ni-job-tuple-icon-srp-rupee sal'
soup5 = soup.find_all('span',class_ = 'ni-job-tuple-icon ni-job-tuple-icon-srp-rupee sal')
```

```python
list5 = [i.text for i in soup5]
```

# **Creating DataFrame:**
* The create_dataframe function takes the lists of job-related information and returns a DataFrame.

```python
import pandas as pd

def create_dataframe(titles, companies, experiences, locations, salaries):
    # Find the minimum length among all lists
    min_length = min(len(titles), len(companies), len(experiences), len(locations), len(salaries))

    # Use zip to iterate over lists simultaneously and truncate or pad as needed
    data = {
        'Job Title': titles[:min_length],
        'Company Name': companies[:min_length],
        'Experience': experiences[:min_length],
        'Location': locations[:min_length],
        'Salary': salaries[:min_length]
    }

    # Create DataFrame
    df = pd.DataFrame(data)
    return df

# Creating DataFrame
create_dataframe(list1, list2, list3, list4, list5)
```

# **Data Visualization:**

## **Visualization 1:** Number of Jobs per Company (Bar Chart)
* the code visualizes the distribution of jobs across different companies by creating a bar chart where each bar represents a company, and the height of the bar corresponds to the number of jobs available at that company. The chart provides a quick overview of the companies with the most job listings.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count the number of jobs per company
company_counts = df['Company Name'].value_counts()

# Plot a bar chart using seaborn
company_counts.plot(kind='bar', color='skyblue')

# Set plot title and labels
plt.title('Number of Jobs per Company')
plt.xlabel('Company Name')
plt.ylabel('Number of Jobs')

# Display the plot
plt.show()
```

## **Visualization 2: Distribution of Experience (Histogram)**
* the code generates a histogram to illustrate the distribution of experience levels among the data science job listings. The green bars represent the frequency of jobs within different experience ranges, and the KDE plot provides a smooth estimate of the underlying probability density.

```python
# Set the figure size
plt.figure(figsize=(10, 8))

# Create a histogram using seaborn
sns.histplot(df['Experience'], bins=20, color='green', kde=True)

# Set the title and labels
plt.title('Distribution of Experience')
plt.xlabel('Experience')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

## **Visualization 3: Count Plot (Distribution of Job Titles)**
* This line displays the created count plot with the specified figure size, providing a visual representation of the distribution of job titles in the data science job listings. The count plot helps visualize the frequency of each job title, aiding in understanding the job distribution.

```python
# Set the figure size
plt.figure(figsize=(10, 8))

# Count plot
sns.countplot(y='Job Title', data=df, order=df['Job Title'].value_counts().index, palette='viridis')

# Add title and labels
plt.title('Distribution of Job Titles')
plt.xlabel('Number of Jobs')
plt.ylabel('Job Title')

# Show the plot
plt.show()
```
